```julia
using Polynomials, Plots
```

```julia
function main()
    
    
    xlin = range(-5, 5, length = 200)
    
    # Creating polynomials
    
    p = Polynomial([1, 3, 5]) # From coefficients
    q = fromroots([1, -4, 4]) # From roots
    
    println(p)
    println(q)
    
    # Math operations
    
    println(p(2))
    println(p + q)
    println(2*q - p)
    
    # Derivatives and integrals
    
    println(derivative(p))
    println(integrate(q))
    
    # Roots
    
    println(roots(q))
    
    # Extrema
    
    xq = roots(q)
    qprime = derivative(q)
    xqq = roots(qprime)
    
    plot(xlin, q.(xlin), label = string(q), legend = :bottomright ,framestyle = :zerolines, dpi = 150)
    plot!(xlin, qprime.(xlin), label = string(qprime))
    scatter!(xq, q.(xq), label = "roots")
    scatter!(xqq, q.(xqq), label = "extrema")
    
end
```

```julia
main()
```
